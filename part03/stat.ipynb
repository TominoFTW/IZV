```python
import pandas as pd
import numpy as np
import scipy.stats as ss
```

# Hypoteza 1

*H0: Na silnicích první třídy se při nehodách umíralo se stejnou pravděpodobností jako nasilnicích třetí třídy*

```python
df = pd.read_pickle("accidents.pkl.gz")
```

Vytvorime novy dataframe, ktery bude obsahovat data pouze ze silnic prvni a treti tridy a nasledne vytvorime sloupec "nasledek", ktery kdyz je hodnota True, tak znamena ze nekdo zemrel, 0 pokud ne

```python
data = df[df["p36"].isin([1,3])].copy()
data["nasledek"] = data["p13a"] > 0

data
```

Vytvorime kontingencni tabulku

```python
table = pd.crosstab(data["nasledek"], data["p36"])
table
```

Pomoci funkce chi2_contingency zjistime p-hodnotu

```python
result = ss.chi2_contingency(table)

result[1]
```

### Zaver 1

Hodnota p-value je mensi nez 0.05, takze zamitame nulovou hypotezu a zavadime alternativni hypotezu, ze je mezi nasledky nehody a tridou silnice statisticky vyznamna spojitost


# Hypoteza 2
*H0: Při nehodách vozidel značky Škoda je škoda na vozidle nižší než při nehodách vozidel Audi*

Vytvorime si novy dataframe, ktery bude obsahovat data pouze o nehodach vozidel znacky Audi a Skoda

```python
data = df[df["p45a"].isin([2,39])]

#give me data set, with only p45a(which is only 2 and 39) and p53 columns
data = data[["p45a", "p53"]]
data
```

Zmenime nazev sloupce

```python
data.set_index("p45a", inplace=True)
data
```

Rozdelime na 2 dataframy a vypocitame jejich mediany a porovname

```python
audi = data.loc[2, "p53"]
skoda = data.loc[39, "p53"]

audi.median(), skoda.median()
```

Spocitame vysledek u-testu

```python
ss.mannwhitneyu(audi, skoda)
```

### Zaver 2
Zde vidime, ze hodnota testu je mensi jak 0.05. Coz potvrzuje nasi hypotezu, ze skoda na vozidlech Skoda je mensi nez u vozidel Audi.
